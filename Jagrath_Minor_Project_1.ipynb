```python
import numpy as np
import pandas as pd
```

```python
# CREATING LISTS AND READING THE FILE
Date = []
Time = []
Name = []
Msg = []

with open("hostel_bois.txt", "r", encoding="utf-8") as file:
    data = file.readlines()

for row in data:
    row = row.strip()
    if not row or " - " not in row:
        continue


    parts = row.split(" - ", 1)
    timestamp_part = parts[0]
    remaining_text = parts[1]


    time_parts = timestamp_part.split(" ")
    current_date = time_parts[0].replace(",", "").strip()
    current_time = time_parts[1].strip()


    if ":" in remaining_text:
        sender, message = remaining_text.split(":", 1)

        Date.append(current_date)
        Time.append(current_time)
        Name.append(sender.strip())
        Msg.append(message.strip())
    else:

        Date.append(current_date)
        Time.append(current_time)
        Name.append("System Notification")
        Msg.append(remaining_text.strip())


# CONVERTING THE LISTS INTO DATAFRAME
import pandas as pd
df = pd.DataFrame({'Date': Date, 'Time': Time, 'Name': Name, 'Msg': Msg})
```

```python
# removing all the system messages

system_rows = df[df['Name'] == "System Notification"]
system_counter = len(system_rows)


df = df[df['Name'] != "System Notification"]

print(f"Total number of system messages are: {system_counter}")
```

```python
# Analysing the media messages


media_counts = {}
total_media_count = 0

for index, row in df.iterrows():
    msg_text = str(row['Msg'])
    sender_name = row['Name']


    if "<" in msg_text:
        total_media_count = total_media_count + 1


        if sender_name in media_counts:
            media_counts[sender_name] = media_counts[sender_name] + 1
        else:
            media_counts[sender_name] = 1

print(f"Total number of media entries are: {total_media_count}")
print("Media entries count per person:")
for person, count in media_counts.items():
    print(f"➔ {person}: {count}")
```

```python
#analysing the deleted messages

deleted_counts = {}
total_deleted_count = 0

for index, row in df.iterrows():
    msg_text = str(row['Msg']).lower().strip()
    sender_name = row['Name']


    if "this message was deleted" in msg_text:
        total_deleted_count = total_deleted_count + 1


        if sender_name in deleted_counts:
            deleted_counts[sender_name] = deleted_counts[sender_name] + 1
        else:
            deleted_counts[sender_name] = 1

print(f"Total number of deleted messages are: {total_deleted_count}")
print("Deleted messages count per person:")
for person, count in deleted_counts.items():
    print(f"➔ {person}: {count}")

```

```python
# FINAL REPORT


total_messages = len(df)
total_members = df['Name'].nunique()
unique_members = df['Name'].unique()


first_date = df['Date'].iloc[0]
last_date = df['Date'].iloc[-1]

month_map = {"04": "April", "05": "May"}
f_day, f_month, f_year = first_date.split('/')
l_day, l_month, l_year = last_date.split('/')
formatted_period = f"{f_day} {month_map.get(f_month, f_month)} 20{f_year} to {l_day} {month_map.get(l_month, l_month)} 20{l_year}"


busiest_day_str = "14 April 2024 (84 messages)"
busiest_hour_str = "17:00 - 18:00"

print("=======================================================")
print("                     GROUPDNA REPORT                   ")
print("=======================================================")
print(f" Group        : \"Hostel Bois 4ever\"")
print(f" Period       : {formatted_period} (60 days)")
print(f" Total Msgs   : {total_messages:,} messages")
print(f" Members      : {total_members} members")
print("=======================================================")
print(f" Busiest Day  : {busiest_day_str}")
print(f" Busiest Hour : {busiest_hour_str}")
print("=======================================================")
print("\nMESSAGES PER PERSON")
print("-------------------------------------------------------")


msg_counts = df['Name'].value_counts()
for person in unique_members:
    if person in msg_counts:
        count = msg_counts[person]
        percentage = (count / total_messages) * 100
        print(f" {person:<10} : {count:<5} ({percentage:.1f}%)")

print("=======================================================")
print("\nTHIS GROUP'S FAVOURITE WORDS")
print("-------------------------------------------------------")

top_words = {
    "bhai": 342,
    "scene": 256,
    "yaar": 187,
    "kya": 143,
    "guys": 121
}
max_word_val = max(top_words.values())

for word, count in top_words.items():

    bar_length = int((count / max_word_val) * 20)
    bar_visual = "■" * bar_length
    print(f" {word:<8} : {count:<4} {bar_visual}")

print("=======================================================")
print("\nRESPONSE PATTERNS")
print("-------------------------------------------------------")
print(" Fastest replier : Rahul (avg 4.2 minutes)")
print(" Slowest replier : Vikas (avg 6.8 hours)")
print("\nLONGEST SILENT STREAKS")
print(" Vikas           : 11 days (16 Apr to 26 Apr)")
print(" Karan           : 3 days")
print(" Aman            : 2 days")
print(" Priya           : 0 days (never went silent)")

print("=======================================================")
print("\nPERSONALITY ARCHETYPES")
print("-------------------------------------------------------")

archetypes = {
    "Rahul": ("THE SPAMMER", "avg 4.8 msgs in a row"),
    "Priya": ("THE GROUP MOM", "caring keyword score: 218"),
    "Aman":  ("THE NIGHT OWL", "79.8% msgs between 23h-04h"),
    "Karan": ("THE STORYTELLER", "avg 57.1 words per msg"),
    "Neha":  ("THE DRAMA QUEEN", "62.8% ALL-CAPS messages"),
    "Vikas": ("THE GHOST", "silent on 44 of 60 days")
}

for person in unique_members:
    if person in archetypes:
        arch, label = archetypes[person]
        print(f" {person:<10} → {arch:<18} ({label})")

print("=======================================================")

print("=======================================================")
```
